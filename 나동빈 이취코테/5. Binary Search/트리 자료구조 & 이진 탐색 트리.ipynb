### 트리 자료구조



- 이진 탐색의 전제 조건 : <b>데이터 정렬</b>
> - 데이터베이스는 내부적으로 **대용량 데이터 처리에 적합한 Tree 자료구조**를 이용하여 항상 데이터가 정렬되어 있음
> - 따라서 데이터베이스에서의 탐색은 (이진 탐색과는 조금 다르지만) 이진 탐색과 유사한 방법을 이용해 탐색을 항상 빠르게 수행하도록 설계되어 있어서 데이터가 많아도 탐색하는 속도가 빠른 것

- 트리 자료구조 : 노드 & 노드의 연결로 표현
    - 노드 : 정보의 단위로서 어떠한 정보를 가지고 있는 개체
    - 주요 특징
        - 트리는 **부모 노드와 자식 노드의 관계**로 표현됨
        - 트리의 최상단 노드 : **루트(Root) 노드**
        - 트리의 최하단 노드 : **단말(Leaf) 노드**
        - 트리에서 일부를 떼어내도 트리 구조이며, 이를 **서브 트리(Subtree)**라고 함
        - 트리는 파일 시스템과 같이 **계층적이고 정렬된 데이터**를 다루기에 적합함
    - 트리 자료구조는 그래프 자료구조의 일종으로 **데이터베이스 시스템, 파일 시스템**과 같은 곳에서 **많은 양의 데이터를 관리하기 위한 목적**으로 사용
    - 큰 데이터를 처리하는 소프트웨어는 대부분 데이터를 **트리 자료구조로 저장**해서 **이진 탐색과 같은 탐색 기법을 이용해 빠르게 탐색**이 가능함

### 이진 탐색 트리



- 이진 탐색 트리 :
    - 트리 자료구조 중 가장 간단한 형태
    - 이진 탐색이 동작할 수 있도록 고안된, 효율적인 탐색이 가능한 자료구조
    - 모든 트리가 다 이진 탐색 트리는 아니며, **왼쪽 자식 노드 < 부모 노드 < 오른쪽 자식 노드**가 성립해야 함
    - 이진 탐색 트리에 데이터를 넣고 빼는 방법은 **알고리즘보다는 자료구조에 가까움**
- 이진 탐색 과정 : 
    1. **루트(Root) 노드**부터 방문하여 찾고자 하는 값(target)과 비교
        1. 찾고자 하는 값(target)이 더 크면, 오른쪽 자식 노드를 방문
        2. 찾고자 하는 값(target)이 더 작으면, 왼쪽 자식 노드를 방문
    2. 방문한 노드를 **부모 노드**로 하여 찾고자 하는 값(target)과 비교
        1. 찾고자 하는 값(target)이 더 크면, 오른쪽 자식 노드를 방문
        2. 찾고자 하는 값(target)이 더 작으면, 왼쪽 자식 노드를 방문
    3. 1 ~ 2를 반복하며 찾고자 하는 원소 값을 찾거나, 자식 노드가 더 이상 존재하지 않으면(return None; 이진 탐색 트리에 원소가 없는 것) 종료
